```python
import numpy as np
import random
import math
import time
```

```python
def function(vector,i):
    temp = np.float64(0)
    for j in range(i):
        temp = temp + vector[j]
    return temp*temp
```

```python
def evalfunction(vector):
    global fes
    result = np.float64(0)
    for i in range(dim):
        result = result + function(vector,i)
    fes = fes + 1
    return result
```

```python
def sort(vector):
    temp = 0 
    for i in range(groupsize):
        if vector[temp] > vector[i]:
            temp = i
    return temp
```

```python
def makerand():
    for i in range(1000000):  
        b = random.randint(0, 99)
        b = np.float64(b/100);
        random_array[i] = b;
        #print(f'{random_array}') 
```

```python
def randf():
    global random_i
    global rands
    if random_i >= 999999:
        random_i = 0
    else:
        random_i = random_i  + 1
    rands = rands + 1
    return random_array[random_i]
```

```python
def exceedsprocess():
    absg = np.float64(0)
    for i in range(groupsize):
        for j in range(dim):
            if generation[j][i][1]+generation[j][i][0]>bond[1]:
                generation[j][i][1]=bond[1]-generation[j][i][0]
            if generation[j][i][1]+generation[j][i][0]<bond[0]:
                generation[j][i][1]=bond[0]-generation[j][i][0]
            # นทนท's Theory!    
            absg=absg+(generation[j][i][1]*generation[j][i][1])
        absg=math.sqrt(absg)
        if absg > maxvelo:
            for j in range(dim):
                generation[j][i][1]=maxvelo*generation[j][i][1]/absg        
```

```python
def PSOgeneration():
    B = 0
    vector = np.empty(dim, dtype=np.float64)
    for i in range(groupsize):
        for j in range(dim):
            generation[j][i][1]=IW*generation[j][i][1]+2*randf()*(generation[j][i][2]-generation[j][i][0])+2*randf()*(leader[j][1]-generation[j][i][0])
    exceedsprocess()
    for i in range(groupsize):
        for j in range(dim):
            generation[j][i][0]=generation[j][i][0]+generation[j][i][1]
    for i in range(groupsize):
        for j in range(dim):
            vector[j]=generation[j][i][0]
        compare[0][i]=evalfunction(vector)
        if compare[0][i]<compare[1][i]:
            compare[1][i]=compare[0][i]
            for j in range(dim):
                generation[j][i][2]=generation[j][i][0]
    B=sort(compare[1])
    leader[1][2]=leader[0][2]
    
    if leader[2][2]>compare[1][B]:
        leader[2][2]=compare[1][B]
        leader[0][2]=compare[1][B]
        for j in range(dim):
            leader[j][1]=generation[j][B][2]
            leader[j][0]=generation[j][B][2]
    else:
        if leader[0][2]>compare[1][B]:
            leader[0][2]=compare[1][B]
            for j in range(dim):
                leader[j][0]=generation[j][B][2]    
    return 0
```

```python
def initialising():
    previous=randf()
    B = 0 
    F = np.empty(groupsize, dtype=np.float64)
    vector = np.empty(dim, dtype=np.float64)
    for i in range(groupsize):
        for j in range(dim):
            generation[j][i][0]=4*previous*(1-previous)
            previous=generation[j][i][0]
            generation[j][i][2]=generation[j][i][0]*(bond[1]-bond[0])+bond[0]
            generation[j][i][0]=generation[j][i][2]
        
    for i in range(groupsize):
        for j in range(dim):
            vector[j]=generation[j][i][0]
        F[i]=evalfunction(vector)   
    B = sort(F)
    
    for i in range(dim):
        leader[i][0]=generation[i][B][0]
        leader[i][1]=generation[i][B][0]
    leader[0][2]=leader[1][2]=leader[2][2]=F[B]
```

```python
#if __name__ == "__main__":
#random::::::::::::begin
random_array = np.empty(1000000, dtype=np.float32)
#random::::::::::::end

#configure::::::::begin
dim = 30
groupsize = 20
T = 3
lifespan = 60

IW = 0.9
maxvelo = 2
bond = [-100,100]
DIM = np.float64(dim)
pi = 3.1415926535897932384626433
#configure::::::::end


#memory::::::::::::begin
generation = np.empty([dim, groupsize, 3], dtype=np.float64)
leader = np.empty([dim, 3], dtype=np.float64)
compare = np.empty([2,groupsize], dtype=np.float64)
 #memory::::::::::::end

#statistic:::::::::begin
fes=0
rands=0
random_i=0
#statistic:::::::::end


op = time.process_time()
makerand()
initialising()
for i in range(2):
    for j in range(groupsize):
        compare[i][j]=1e200
while(1):
    PSOgeneration()
    print(f'fes:{fes},_____leader:{leader[2][2]}, _____IW:{IW}')
    IW = (0.5*(1-np.float64(fes/200000)))+0.4
    if fes>200000 :
        ed =  time.process_time()    
        print(f'EFS={fes}_____{rands}')
        print(f'execution time: {ed-op}')
        break
```
